```python
import requests
import pandas as pd
import os


dataverse_base_url = "https://dataverse.harvard.edu/api/datasets/:persistentId/"
download_base_url = "https://dataverse.harvard.edu/api/access/datafile/"


dataset_doi = "doi:10.7910/DVN/EGZHFV"


download_directory = "downloaded_files"


os.makedirs(download_directory, exist_ok=True)


def download_and_convert_to_csv(file_id, file_name):

    file_url = f"{download_base_url}{file_id}"
    response = requests.get(file_url, stream=True)
    if response.status_code == 200:
        txt_file_path = os.path.join(download_directory, f"{file_name}.txt")


        with open(txt_file_path, "wb") as f:
            f.write(response.content)
        print(f"Downloaded {txt_file_path}")


        try:

            df = pd.read_csv(txt_file_path, sep='\t')
            csv_file_path = os.path.join(download_directory, f"{file_name}.csv")
            df.to_csv(csv_file_path, index=False)
            print(f"Converted {txt_file_path} to {csv_file_path}")
        except Exception as e:
            print(f"Failed to convert {txt_file_path} to CSV: {e}")
    else:
        print(f"Failed to download {file_name}")


dataset_url = f"{dataverse_base_url}?persistentId={dataset_doi}"
response = requests.get(dataset_url)


if response.status_code == 200:
    files_info = response.json().get("data", {}).get("latestVersion", {}).get("files", [])


    for file_info in files_info:
        file_id = file_info["dataFile"]["id"]
        file_name = file_info["dataFile"]["filename"].replace('.txt', '')
        download_and_convert_to_csv(file_id, file_name)
else:
    print("Failed to retrieve dataset information.")
```

```python
import shutil


source_dir = '/content/downloaded_files'
destination_dir = '/content/drive/My Drive/IMDEA'


os.makedirs(destination_dir, exist_ok=True)


for filename in os.listdir(source_dir):
    source_path = os.path.join(source_dir, filename)

    if os.path.isfile(source_path):
        shutil.copy(source_path, destination_dir)
        print(f"Copied {filename} to {destination_dir}")
```
